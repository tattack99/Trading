```python
# Getting the data historical data from eurusd
r = requests.get(market_url_eurusd_hist , headers = payload)
print(r.status_code)
print(r.json())
```

```python
# Adding the history data to a data frame
eurusd_hist = r.json()
df = pd.DataFrame(eurusd_hist["prices"])
print(df)
df.set_index("snapshotTime", inplace = True)
print(df)
```

```python

#### Correct way to get values from asian session####
df_asia = df.loc[asia_session_open_time:asia_session_close_time]
print(df_asia)
#### Change the code above to get time for NY or London
```

```python
####### Getting the ask / sell max during asian session ######
asian_ask = []
for time in df_asia.index:
    print(df.loc[time , "highPrice"]["ask"])
    asian_ask.append(df.loc[time, "highPrice"]["ask"])
print(asian_ask)
df_asian_ask = pd.DataFrame(asian_ask)
print(df_asian_ask.max())
```

```python
####### Getting the bid / buy max during asian session ######
asian_bid = []
for time in df_asia.index:
    print(df.loc[time , "lowPrice"]["ask"])
    asian_bid.append(df.loc[time, "lowPrice"]["ask"])
print(asian_bid)
df_asian_bid = pd.DataFrame(asian_bid)
print(df_asian_bid.max())
```

```python
# Printing out max and min value for asian session buy / sell
print("Asian session, highest sell postion: " + str(df_asian_ask.max()))
print("Asian session, lowest sell postion: " + str(df_asian_ask.min()))
print("\n")
print("Asian session, highest buy postion: " + str(df_asian_bid.max()))
print("Asian session, lowest buy postion: " + str(df_asian_bid.min()))
```
