```python
### COT NET POSITION ###

# List of all the parties
commercial_cot_net = []
noncommercial_cot_net = []
nonrept_cot_net = []

#List of the date
today_list = []

# Setting up a for loop to look through data and calculate net position for all parties
for today in cot_data_index.index:
    
    # Appending today to list
    today_list.append(today)
    
    # Commercials
    today_cot_long = cot_data_index.loc[today, ["Comm_Positions_Long_All"]]
    today_cot_short = cot_data_index.loc[today, ["Comm_Positions_Short_All"]]  
    commercial_cot_net.append(int(today_cot_long) - int(today_cot_short) )
    
    # Non - Commercials
    today_cot_long = cot_data_index.loc[today, ["NonComm_Positions_Long_All"]]
    today_cot_short = cot_data_index.loc[today, ["NonComm_Positions_Short_All"]]  
    noncommercial_cot_net.append(int(today_cot_long) - int(today_cot_short))
    
    # Non - Reportable
    today_cot_long = cot_data_index.loc[today, ["NonRept_Positions_Long_All"]]
    today_cot_short = cot_data_index.loc[today, ["NonRept_Positions_Short_All"]]  
    nonrept_cot_net.append(int(today_cot_long) - int(today_cot_short))

# Creating dictionary to hold all data
net_cot_data = {"date" : today_list,
                "commercials_net" : commercial_cot_net,
                "non_commercials_net" : noncommercial_cot_net,
                "non_rept_net" : nonrept_cot_net}

# Change date to index
net_cot_data = pd.DataFrame(data = net_cot_data, index = today_list)
net_cot_data = net_cot_data.set_index("date")

# Exporting dataframe 
net_cot_data.to_csv(r'C:\Users\timmo\Google Drive\1.Trading\4.Data\COT\net_cot_data_eurusd.csv',  mode = "a")

```

```python
### COT INDEX ### 

# Getting the data
cot_data_net = pd.read_csv(r'C:\Users\timmo\Google Drive\1.Trading\4.Data\COT\net_cot_data_eurusd.csv' , parse_dates = ["date"])

# Setting dates as index
cot_data_net.set_index("date" , inplace = True)

# Setting time delta of data 3 years ago
tdelta = timedelta(days = -1095)

# Lists
today_list = []
cot_index_list = []

for today in cot_data_net.index:
    
    # Converting today string to datetime
    today_three_year_ago = today + tdelta
    
    # Commercial cot index
    
    print("\n")

    # Commercial net today 
    com_cot_net_today = cot_data_net.loc[today, ["commercials_net"]]
    print("Commercials today net: ")
    print(com_cot_net_today)
    
    
    # Commercial net 3 years ago max and min
        
    print("\n")
    
    # Check validity of old period
    try:
        cot_data_net.loc[today_three_year_ago, ["commercials_net"]]
    except KeyError:
        print("No value for this time priod. ")
        
    # If validity ok then check for max and min
    else:
        # Max
        com_cot_net_old_max = cot_data_net.loc[today_three_year_ago:today, ["commercials_net"]].max()
        print("Commercial max past 3 years")
        print(com_cot_net_old_max)
        
        # Min
        com_cot_net_old_min = cot_data_net.loc[today_three_year_ago:today, ["commercials_net"]].min()
        print("Commercial min past 3 years")
        print(com_cot_net_old_min)
        
        # Calculate cot index
        cot_index = ((int(com_cot_net_today) - int(com_cot_net_old_min))/(int(com_cot_net_old_max) - int(com_cot_net_old_min)))*100
       
        # Appending cot index and date into a list
        cot_index_list.append(cot_index)
        today_list.append(today)
        
        print("\n")
        print("COT INDEX: ")
        print(cot_index)

        
# Putting data into a dataframe and exporting it to csv
cot_index_data = {"date" : today_list, "cot_index" : cot_index_list}

cot_index_data = pd.DataFrame(data = cot_index_data, index = today_list)
cot_index_data = cot_index_data.set_index("date")

# Exporting data frame
cot_index_data.to_csv(r'C:\Users\timmo\Google Drive\1.Trading\4.Data\COT\cot_index_eurusud.csv', mode = "a")
```
